```python
# Paso 1: generar los códigos de temporada
temporadas = []

for anio in range(2000, 2026):
    inicio = str(anio)[2:]   # últimos 2 dígitos del año inicio
    fin = str(anio + 1)[2:]  # últimos 2 dígitos del año fin
    codigo = inicio + fin
    temporadas.append(codigo)

print(temporadas)
```

```python
import requests
import os

# Paso 2: crear carpeta donde guardar los archivos
carpeta = r"c:\Users\brger\OneDrive\Desktop\Project\Football_Prediction\data\raw"
os.makedirs(carpeta, exist_ok=True) 

# Paso 3: descargar cada CSV
for codigo in temporadas:
    url = f"https://www.football-data.co.uk/mmz4281/{codigo}/SP1.csv"
    nombre_archivo = f"{carpeta}/SP1_{codigo}.csv"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        with open(nombre_archivo, "wb") as f:
            f.write(response.content)
        print(f"✅ Descargado: SP1_{codigo}.csv")
    else:
        print(f"❌ Error en temporada {codigo} - Status: {response.status_code}")
```
